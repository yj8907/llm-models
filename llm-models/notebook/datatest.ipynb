```python
import tiktoken
import json
import base64

def export_gpt4_tokenizer_for_megatron(output_path="gpt4_tokenizer.json"):
    """
    Export GPT-4 tokenizer vocabulary in Megatron-compatible format.
    
    Expected format by Megatron:
    [
        {"rank": 0, "token_bytes": "base64_encoded_bytes", "token_str": "utf8_string"},
        {"rank": 1, "token_bytes": "base64_encoded_bytes", "token_str": "utf8_string"},
        ...
    ]
    """
    # Get the GPT-4 encoding (cl100k_base)
    enc = tiktoken.encoding_for_model("gpt-4")
    
    # Get mergeable ranks: {bytes: rank}
    mergeable_ranks = enc._mergeable_ranks
    
    # Create list sorted by rank
    vocab_list = []
    for token_bytes, rank in sorted(mergeable_ranks.items(), key=lambda x: x[1]):
        vocab_list.append({
            "rank": rank,
            "token_bytes": base64.b64encode(token_bytes).decode("utf-8"),
            "token_str": token_bytes.decode("utf-8", errors="replace")
        })
    
    # Verify the vocabulary is complete and sequential
    assert len(vocab_list) == len(mergeable_ranks)
    assert all(vocab_list[i]["rank"] == i for i in range(len(vocab_list)))
    
    # Save to JSON
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(vocab_list, f, ensure_ascii=False, indent=2)
    
    print(f"Success! Saved tokenizer to {output_path}")
    print(f"Vocabulary size: {len(vocab_list)}")
    print(f"First 3 tokens:")
    for i in range(min(3, len(vocab_list))):
        print(f"  Rank {vocab_list[i]['rank']}: {repr(vocab_list[i]['token_str'])}")
```

```python
export_gpt4_tokenizer_for_megatron('/home/ubuntu/projects/llm-models/llm-models/dataset/work/dclm/megatron/gpt-4.json')
```
